**1) Write a Python function to calculate the factorial of a number**



```python
def fact(x):
  num=1
  for i in range(1,x+1):
    num=num*i
  print(num)

fact(7)
```

**2) Write a Python function to check whether a number is in a given range. Hint: specify a range first
in your code then check whether that number is in your range or not**

```python
def num():
  n=int(input("Enter a number:"))
  for i in range(1,101):
    if n//i==1 and n%i==0:
      print("Yes , this number exists within 1-100 range ")
      break
  else:
    print("No, it doesnt exist within 1-100 range")
num()
```

**3) Write a Python function to multiply all the numbers given in the input. E.g.: Input numbers 1,3,5.
Output: 15**

```python
def mult():
  y=1
  x=str(input("Enter a number:"))
  for i in x:
    y=y*int(i)
  print(y)
mult()
```
