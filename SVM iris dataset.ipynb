```python
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
iris = datasets.load_iris()
```

```python
print("Features: ", iris.feature_names)
```

```python
print("Labels: ", iris.target_names)
```

```python
iris.data.shape
```

```python
print(iris.data[0:5])
```

```python
print(iris.target)
```

```python
X = iris.data[:, :2]
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.80, test_size=0.20, random_state=101)
```

```python
rbf = svm.SVC(kernel='rbf', gamma=0.5, C=0.1).fit(X_train, y_train)
poly = svm.SVC(kernel='poly', degree=3, C=1).fit(X_train, y_train)
```

```python
poly_pred = poly.predict(X_test)
rbf_pred = rbf.predict(X_test)
```

```python
poly_accuracy = metrics.accuracy_score(y_test, poly_pred)
print('Accuracy (Polynomial Kernel): ', "%.2f" % (poly_accuracy*100))
```

```python
rbf_accuracy = metrics.accuracy_score(y_test, rbf_pred)
print('Accuracy (RBF Kernel): ', "%.2f" % (rbf_accuracy*100))
```
