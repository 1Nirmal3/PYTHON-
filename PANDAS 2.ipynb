```python
df=df.drop_duplicates(subset=['App'])
df.duplicated(subset=['App']).sum()
```

```python
print(df.Category.value_counts())
```

```python
df['Category'].replace("1.9","FAMILY",inplace=True)
```

```python
print(df.Category.value_counts())
```

```python
df['Rating'].fillna(df['Rating'].mean(),inplace=True)
```

```python
df['Rating'].isna().sum()
```

```python
df.info()
```

```python
df['Size']=df["Size"].str.replace('Varies with device','200000')
df['Size']=df['Size'].str.replace('M','000000')
df['Size']=df['Size'].str.replace('K','000')
df['Size']=pd.to_numeric(df['Size'],errors='coerce')
df['Size']=df['Size'].fillna(df['Size'].mean())
```

```python
df.info()
```

```python
df['Installs']=df['Installs'].str.replace(',','')
df['Installs']=df['Installs'].str.replace('+','')
df['Installs']=df['Installs'].str.replace('Free','')
```

```python
df.head()
```

```python
print(df.Installs.value_counts())
```

```python
print(df.Type.value_counts())
```

```python
df['Installs']=pd.to_numeric(df['Installs'])
df['Installs']=df['Installs'].fillna(df['Installs'].mean())
```

```python
df['Type']=df['Type'].str.replace('0','Free')
```

```python
print(df.Type.value_counts())
```

```python
print(df.Price.value_counts())
```

```python
df['Price']=df['Price'].str.replace('$','')
df['Price']=pd.to_numeric(df['Price'],errors='coerce')
df['Price']=df['Price'].fillna(df['Price'].mean())
```
