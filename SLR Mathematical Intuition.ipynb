```python
x=[1,2,3,4,5]
y=[3,4,2,4,5]
```

```python
def mean(l):
  mean=0
  count=0
  for i in l:
    mean=mean+i
    count=count+1
  mean=mean/count
  return mean
```

```python
mean(x)
```

```python
mean(y)
```

```python
import pandas as pd
df=pd.DataFrame()
```

```python
df['x']=x
df['y']=y
```

```python
df
```

```python
def imean(x):
  l=[]
  for i in x:
    a=i-(mean(x))
    l.append(a)
  return l
    
```

```python
imean(x)
```

```python
df['x-xmean']=imean(x)
```

```python
imean(y)
```

```python
df['y-ymean']=imean(y)
```

```python
df
```

```python
def imeansquare(x):
  l=[]
  for i in x:
    a=((i-mean(x))**2)
    l.append(a)
  return l
```

```python
imeansquare(x)
```

```python
df['(x-xmean)^2']=imeansquare(x)
```

```python
df['(x-xmean)(y-ymean)']=df['x-xmean']*df['y-ymean']
```

```python
df.loc["Total"] = df.sum()
```

```python
df
```

```python
m=df['(x-xmean)(y-ymean)']['Total']/df['(x-xmean)^2']['Total']
```

```python
m
```

```python
c=mean(y)-(m*mean(x))
```

```python
c
```

```python
y_pred=[]
for i in df['x'].iloc[:5,]:
  num=(m*i)+c
  y_pred.append(num)

```

```python
sum=0
for i in y_pred:
  sum=sum+i
```

```python
y_pred.append(sum)
```

```python
df['y_pred']=y_pred
```

```python
df
```

```python
df['y_pred-ymean']=df['y_pred']-mean(y)
```
