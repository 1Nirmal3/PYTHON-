**3)Write a program to find the area of a triangle. Ask the user to input three sides 
of a triangle and then calculate the area. 
If a, b and c are three sides of a triangle. Then,
s = (a+b+c)/2
area = ‚àö(s(s-a)(s-b)(s-c))**

```python
a=int(input("Enter length of first side:"))
b=int(input("Enter length of second side:"))
c=int(input("Enter length of third side:"))
s=(a+b+c)/2
area=(s*(s-a)*(s-b)*(s-c))**(1/2)
print(area)
```

**7)Write a program to calculate compound interest
A = P(1 + R/100)t
Compound Interest = A ‚Äì P
Where,
A is amount
P is principle amount
R is the rate and
T is the time span**


```python
P=int(input("Enter Principle Amount:"))
R=float(input("Enter interest Rate:"))
T=int(input("Enter Time Period:"))
A=P*(1+(R/100))*T
CI=A-P
print("Compound Interest:",CI)
```

**9)Body Mass Index (BMI) is a measure of health based on weight. It can be 
calculated by taking your weight in kilograms and dividing it by the square of 
your height in meters. Write a program that prompts the user to enter the weight 
in pounds and height in inches and display the BMI. Note that one pound is 
0.45359237 kilograms and one inch is 0.0254 meters.**


```python
w=float(input("Enter weight(in pounds):"))
h=int(input("Enter height(in inches):"))
bmi=(w*0.45359237)/((h*0.0254)**2)
print(bmi)
```

**10)Write a program must satisfy the following requirements: 
‚ñ†It must let the user enter the interest rate, the loan amount, and the number of 
years for which payments will be made. 
‚ñ† It must compute and display the monthly payment and total payment amounts.**

```python
r=float(input("Enter Annual Interest Rate:"))
t=int(input("Enter no. of years:"))
p=int(input("Enter principal amount:"))
m_itr= r/1200
m_amount=(p*m_itr)/(1-(1/(1+m_itr))**(t*12))
print("Monthly Payment:",m_amount)
total=m_amount*t*12
print("Total Payment:",total)
```
