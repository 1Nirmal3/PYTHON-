**1. Write a program to check whether the year is leap year or not.**



```python
y=int(input("Enter the total year:"))
if  y % 4 == 0 and (y % 100 != 0 or y % 400 == 0):
  print("It's a leap year")
else:
  print("It's not a leap year")
```

b. 25 to 45 - E
c. 45 to 50 - D
d. 50 to 60 - C
e. 60 to 80 - B
f. Above 80 - A

**6. A student will not be allowed to sit in exam if his/her attendance is less than 75%.Take 
following input from user:
Number of classes held 
Number of classes attended.
and print percentage of class attended and whether student is allowed to sit in exam or 
not.**

```python
c=int(input("Enter total number of classes held:"))
a=int(input("Enter total number of classes attended:"))
p=(a/c)*100
print("Attendance Percentage:",p,"%")
if p<75:
  print("You're attendance doesnt meet the requirements")
else:
  print("You are allowed to attend the exams")
```

**8. Write a Python program to calculate a dog's age in dog's years. 
Note: For the first two years, a dog year is equal to 10.5 human years. After that, each 
dog year equals 4 human years.
Expected Output:
Input a dog's age in human years: 15 
The dog's age in dog's years is 73**

```python
a=float(input("Enter Dog's Age in human years:"))
x=2*10.5
y=(a-2)*4
age=x+y
print("Dog's Age in Dog's Years:",age)
```
